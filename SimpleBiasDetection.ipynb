```python
lodata= pd.DataFrame(lodata)
datam=lodata.groupby(["age"])["loan_approved"].value_counts(normalize=True).unstack()
datam.fillna(0,inplace=True)
datam
```

```python
dgl=lodata.groupby(["age"])["loan_approved"].value_counts(normalize=False).unstack()
dgl.fillna(0, inplace=True)
dgl
```

```python
dg = lodata.groupby(["age","income"])["loan_approved"].all().unstack()
dg.fillna('-', inplace=True)
dg
```
